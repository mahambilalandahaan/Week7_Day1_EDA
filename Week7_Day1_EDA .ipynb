```python
#------EDA(Exploratory Data Analysis)------
# 1. mount dat
import matplotlib.pyplot as plt
from google.colab  import drive
drive.mount('/content/drive')
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/spam.csv',encoding='utf-8')
df.head()
# 2.Exploratory Analysis

df.info()
df['label'].value_counts()
df['label'].value_counts().plot(kind='bar',color='green')

# 3.Basic words frequency count
from collections import Counter
#  combine all messages into a single string
all_words=" ".join(df['text']).lower().split()

#count the 10 most common words
word_counts=Counter(all_words).most_common(10)
word_counts

#4 Top 10 most frequent words visualization

plt.figure(figsize=(10,6))
words , count=zip(*word_counts)
plt.bar(words,count,color='indigo')
plt.xticks(rotation=45)
plt.xlabel('Words')
plt.ylabel('Counts')
plt.title('Top 10 Most Frequent Words')
plt.show()

# 5. Light text cleaning
import re
def clean_text(msg):
  msg=msg.lower()   #lowercase
  msg=re.sub(r'[^\w\s]',' ',msg)
  return msg
df['clean_text']=df['text'].apply(clean_text)
df[['text','clean_text']].head()
```
